```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("D:\college\cw py\DataScienceSalaries_82fcd341-306b-4fb9-b353-f4a4db520374_93472_.csv")
```

```python
df.head(15)
```

```python
df.drop(columns=['salary', 'salary_currency'])
```

```python
nan_count_per_row = df.isna().sum(axis=1)
total_rows_with_nan = (nan_count_per_row > 0).sum()
print("Total rows with NaN values:", total_rows_with_nan)
```

```python
df.dropna(inplace=True)
```

```python
duplicates = df[df.duplicated()]

if duplicates.empty:
    print("No duplicate values found.")
else:
    print("Duplicate values found:")
    print(duplicates)
```

```python
unique_values = df.apply(pd.Series.nunique)

print("Unique values in each column:")
print(unique_values)
```

```python
df.rename(columns={'SE': 'Senior Level/Expert', 'MI': 'Medium Level/Intermediate', 'EN': 'Entry Level', 'EX': 'Executive Level'})
```

```python
sum_value = df['salary_in_usd'].sum()
mean_value = df['salary_in_usd'].mean()
std_dev = df['salary_in_usd'].std()
skewness = df['salary_in_usd'].skew()
kurt = df['salary_in_usd'].kurtosis()
print("Sum:", sum_value)
print("Mean:", mean_value)
print("Standard Deviation:", std_dev)
print("Skewness:", skewness)
print("Kurtosis:", kurt)
```

```python
numeric_columns = df.select_dtypes(include=[np.number]).columns.tolist()
correlation_matrix = df[numeric_columns].corr()
print(correlation_matrix)
```

```python
top_jobs = df['job_title'].value_counts().head(15)
plt.figure(figsize=(10, 6))
top_jobs.plot(kind='bar')
plt.title('Top 15 Jobs')
plt.xlabel('Job Title')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
mean_salary_by_job = df.groupby('job_title')['salary_in_usd'].mean().nlargest(15)
plt.figure(figsize=(10, 6))
mean_salary_by_job.plot(kind='bar')
plt.title('Mean Salary by Job Title (Top 15)')
plt.xlabel('Job Title')
plt.ylabel('Mean Salary (USD)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
mean_salary_by_experience = df.groupby('experience_level')['salary_in_usd'].mean()
plt.figure(figsize=(10, 6))
mean_salary_by_experience.plot(kind='bar')
plt.title('Mean Salary by Experience Level')
plt.xlabel('Experience Level')
plt.ylabel('Mean Salary USD')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['salary_in_usd'], bins= 40, color='skyblue', edgecolor='black')
plt.title('Histogram of ' + 'salary_in_usd')
plt.xlabel('salary_in_usd')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.boxplot([df[df['work_year'] == 2023]['salary_in_usd'],
             df[df['work_year'] == 2022]['salary_in_usd'],
             df[df['work_year'] == 2021]['salary_in_usd'],
             df[df['work_year'] == 2020]['salary_in_usd']],
            labels=['2023', '2022', '2021', '2020'],
            patch_artist=True,  # Fill the boxplot with colors
            boxprops=dict(facecolor='skyblue', color='black'),  # Set box colors
            whiskerprops=dict(color='black'),  # Set whisker colors
            capprops=dict(color='black'),  # Set cap colors
            medianprops=dict(color='black'))  # Set median line color
plt.title('Box Plot of Salary by Work Year', fontsize=16)  # Adding title with adjusted font size
plt.xlabel('Work Year', fontsize=14)  # Adding x-axis label with adjusted font size
plt.ylabel('Salary (USD)', fontsize=14)  # Adding y-axis label with adjusted font size
plt.grid()  # Adding horizontal grid lines with dashed style and transparency
plt.tight_layout()  # Ensuring tight layout to prevent clipping of labels
plt.savefig("boxplot.png")  # Saving the plot as a PNG file
plt.show()
```
